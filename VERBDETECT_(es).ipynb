```python
def detectar_verbos_func(frase):
  frase=frase.lower()
  palabras=frase.split()
  #terminaciones en infinitivo"
  infinitivos=["ar","er","ir"]
  #conjugaciones
  conjugaciones=["o","as","a","amos","ais","an","é","aste","o","amos","asteis","aron","í","iste","ió", "imos","isteis","ieron"]
  noVerbos=["yo","tú","él","ellos","nosotros,","perro","gato","auto","parque","mientras"]
  verbos=[]
  for palabra in palabras:
    if palabra in noVerbos: continue
    if any (palabra.endswith(term)
    for term in infinitivos): verbos.append(palabra+" infinitivo")
    elif any(palabra.endswith(term)
    for term in conjugaciones):
      verbos.append(palabra+" conjugado")
  return verbos
```

```python
#demo
frase="El perro corrió y saltó en el parque mientras yo corro y salto"
verbos_detectados=detectar_verbos_func(frase)
print("verbos detectados")
for verbo in verbos_detectados:
    print("-",verbo)
```

```python
#Usando librerias
import spacy
spacy.cli.download("es_core_news_sm")
nlp=spacy.load("es_core_news_sm")
frase="El perro corrió y saltó en el parque mientras yo corro y salto"
doc=nlp(frase)
verbos=[]
for token in doc:
    if token.pos_=="VERB":
        verbos.append(token.text)
print("verbos detectados")
for verbo in verbos:
    print("-",verbo)
```
